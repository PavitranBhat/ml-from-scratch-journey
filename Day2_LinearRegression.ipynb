Day 2 - Linear Regression

1. What is Linear Regression?
    Linear Regression is a supervised learning algorithm used for predicting continuous values.
    It finds a straight line that best fits the data.
Example:
    Predicting house price based on area.

Mathematically: 𝑦 = 𝑚𝑥 + 𝑏
Where:
y → Predicted value
x → Input feature
m → Slope (weight)
b → Intercept

The model tries to find the best values of m and b.

2. What is the Goal?
Minimize the error between: Actual value and Predicted value

Error formula:
𝐸𝑟𝑟𝑜𝑟 = 𝐴𝑐𝑡𝑢𝑎𝑙 − 𝑃𝑟𝑒𝑑𝑖𝑐𝑡𝑒𝑑

To measure total error, we use:
Cost Function (MSE – Mean Squared Error)
𝑀𝑆𝐸 = 1/𝑛 ∑(𝑦−𝑦^)2

We square errors to:
    Remove negative sign
    Penalize large errors more

3. Gradient Descent
    Imagine you're standing on a hill.
    You want to reach the lowest point.

Gradient Descent: 
    Takes small steps
    Moves in direction of minimum error
    Repeats until cost is lowest

That’s how model learns best m and b.

```python
import numpy as np  
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

Step 2: Create Simple Dataset

```python
# Example: Study hours vs Marks

X = np.array([1, 2, 3, 4, 5]).reshape(-1,1) #.reshape(-1,1) : Scikit-learn expects input data in 2D format
y = np.array([35, 40, 50, 65, 70])
```

Step 3: Train Model

```python
model = LinearRegression()
model.fit(X, y) #fit means learn from X(input),learn from y(output),find best slope and intercept
```

Step 4: Predictions

```python
y_pred = model.predict(X)
 # y= mx+b formula used internally
 # It calculates marks for each study hour y_pred contains predicted values
```

Step 5: Visualize

```python
plt.scatter(X, y) # this draws actual datapoints
plt.plot(X, y_pred) #This draws the regression line(best fit)
plt.xlabel("Study Hours") # label is used for title for clarity
plt.ylabel("Marks")
plt.title("Linear Regression Example")
plt.show() #displays the graph
```

Step 6: Check Parameters

```python
print("Slope:", model.coef_) #coef_ gives slope(m)
print("Intercept:", model.intercept_) #intercept = b 
```
